## **Starbucks Ürünleri Besin Değerleri Veri Seti**

**Açıklama:** Bu veri seti bize Starbucks ürünlerinin besin değerleri hakkında veriler sunar. Ürünlerin kalorisi, yağ, lif ve protein miktarı gibi bilgiler verir.

**Boyut:** Veri setimiz 77 satır ve 11 sütundan oluşuyor.

#### **Ürün Kategorilerinin Açıklaması:**

1. **Bakery (Fırın Ürünleri):** Starbucks'un fırınlanmış ürünlerini içerir. Bu genellikle ekmekler, kruvasanlar, kekler ve diğer tatlı veya tuzlu fırın ürünlerini kapsar.

2. **Bistro Box:** Bistro box, genellikle atıştırmalık veya hafif bir öğün olarak sunulan çeşitli malzemelerin bir araya getirildiği bir kutudur. İçeriği genellikle peynir, meyve, kuruyemiş, kraker ve belki de bir tür protein içerebilir.

3. **Petite:** "Petite" terimi, küçük veya minyatür anlamına gelir. Starbucks'un "Petite" kategorisi genellikle küçük boyutlu tatlılar veya atıştırmalıkları içerir, örneğin minik kekler veya kurabiyeler gibi.

4. **Hot Breakfast (Sıcak Kahvaltı):** Bu kategori, sıcak servis edilen kahvaltı seçeneklerini içerir. Örnekler arasında sandviçler, yumurtalı yemekler ve sıcak tahıllar bulunabilir.

5. **Salad (Salata):** Salata kategorisi, genellikle hafif ve sağlıklı bir seçenek olan çeşitli yeşil yapraklı veya malzemeli salataları içerir.

6. **Sandwich:** Starbucks'un sandviç kategorisi, çeşitli ekmek türleri arasına yerleştirilmiş çeşitli iç malzemeleri içerir. Örneğin, tavuklu, peynirli veya sebzeli sandviçler olabilir.

7. **Parfait:** Bir "parfait", genellikle yoğurt, meyve ve granola gibi katmanlı malzemelerden oluşan bir tatlı veya atıştırmalıktır. Starbucks'ta bu genellikle yoğurt bazlı ve meyve içeren bir seçenek olarak sunulur.

### **Sayısal verilerin korelasyonuna bakalım**

```python
df.head()
```

```python
numeric_columns = df.select_dtypes(include='number')
corr_matrix = numeric_columns.corr()
```

```python
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Bu korelasyon matrisinden çıkarılabilecek bazı analiz sonuçları şunlardır:

1. **Kalori ve Yağ İlişkisi:** Kaloriler ile yağ arasında güçlü bir pozitif korelasyon gözlemlenmektedir (0.76). Bu, yüksek kalorili gıdaların genellikle yüksek yağ içeriğine sahip olduğunu gösterir.

2. **Karbonhidrat ve Yağ İlişkisi:** Karbonhidratlar ile yağ arasında pozitif bir korelasyon bulunmaktadır (0.14), ancak bu korelasyon kalori ile yağ arasındaki kadar güçlü değildir. Bu, karbonhidratlar ve yağın genellikle birlikte tüketildiğini, ancak kalorinin yağdan daha çok karbonhidrat ile ilişkili olduğunu gösterebilir.

3. **Lif ve Diğer Besin İlişkileri:** Lif miktarı ile diğer besin ögeleri arasında genellikle zayıf veya orta düzeyde bir korelasyon bulunmaktadır. Özellikle protein ile lif arasında pozitif bir korelasyon (0.49) dikkat çekicidir, bu da lif açısından zengin gıdaların genellikle yüksek protein içeriğine sahip olduğunu gösterebilir.

4. **Karbonhidrat ve Protein İlişkisi:** Karbonhidratlar ile protein arasında zayıf bir negatif korelasyon gözlemlenmektedir (-0.05). Bu, karbonhidratların protein içeriği ile ters orantılı olduğunu gösterebilir, yani yüksek karbonhidrat içeren gıdaların genellikle düşük protein içeriğine sahip olduğunu gösterir.

Bu analiz sonuçları, bir gıdanın belirli besin ögeleriyle ilişkili olma eğilimini gösterir ve bir gıdanın beslenme profilini anlamak için faydalı bir bakış sağlar. Bu bilgiler, sağlıklı beslenme veya belirli besin ögelerine odaklanan diyet planlaması gibi alanlarda kullanılabilir.

**Starbucks Ürün Besin Değerleri Analizi**

Bu analizde, Starbucks'un menüsündeki ürünlerin besin değerlerine odaklandık. İşte elde ettiğimiz bulgular:

- **Kalori Dağılımı:** Çoğu ürün, 350 ila 400 arası kalori içermektedir. Bu aralıkta yaklaşık 10-15 adet ürün bulunmaktadır. Menüdeki ürünlerin genelde 300 ve üzeri kalori içerdiği gözlemlenmiştir. Düşük kalorili ürünlerin miktarı azdır.

- **Protein İçeriği:** Protein miktarı arttıkça, ürün çeşidi azalmaktadır. Menüdeki ürünlerin büyük çoğunluğu 0-10 arası protein içermektedir.

- **Yağ İçeriği:** Ürünlerin büyük çoğunluğu 5-20 arası yağ içermektedir. Bu da Starbucks menüsündeki ürünlerin yağ oranının ortalama seviyede olduğunu göstermektedir.

- **Karbonhidrat Dağılımı:** Menüdeki ürünlerin çoğunluğu 60 gram kadar karbonhidrat içermektedir. Bununla birlikte, 0-50 gram arası karbonhidrat içeren ürünlerin sayıca birbirine yakın olduğu görülmüştür.

Bu analiz sonucunda, Starbucks'un menüsündeki ürünlerin genel olarak ortalama seviyede kalori, yağ, protein ve karbonhidrat içeriğine sahip olduğu görülmektedir. Daha düşük kalorili ve yüksek proteinli seçeneklerin menüye eklenmesi, çeşitliliği artırabilir ve müşteri tercihlerine daha fazla seçenek sunabilir.
